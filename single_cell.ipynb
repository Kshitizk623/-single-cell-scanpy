```python
# Initialize lists for tumor and normal data
tumor_data = []
normal_data = []

# Load and append data
for folder in tumor_folders:
    adata = sc.read_10x_mtx(os.path.join(data_path, folder), var_names="gene_symbols", cache=True)
    tumor_data.append(adata)

for folder in normal_folders:
    adata = sc.read_10x_mtx(os.path.join(data_path, folder), var_names="gene_symbols", cache=True)
    normal_data.append(adata)

# Concatenate data
tumor_adata = tumor_data[0].concatenate(tumor_data[1:], batch_key="batch")
normal_adata = normal_data[0].concatenate(normal_data[1:], batch_key="batch")

print("Combined tumor data:", tumor_adata)
print("Combined normal data:", normal_adata)
```

```python
# Calculate mitochondrial gene percentage
tumor_adata.var["mt"] = tumor_adata.var_names.str.startswith("MT-")  # Identify mitochondrial genes
tumor_adata.obs["n_genes_by_counts"] = (tumor_adata.X > 0).sum(axis=1)  # Number of genes expressed
tumor_adata.obs["total_counts"] = tumor_adata.X.sum(axis=1)  # Total UMI counts
tumor_adata.obs["pct_counts_mt"] = (
    tumor_adata[:, tumor_adata.var["mt"]].X.sum(axis=1).A1 / tumor_adata.obs["total_counts"].values
) * 100
```

```python
normal_adata.var["mt"] = normal_adata.var_names.str.startswith("MT-")  # Identify mitochondrial genes
normal_adata.obs["n_genes_by_counts"] = (normal_adata.X > 0).sum(axis=1)  # Number of genes expressed
normal_adata.obs["total_counts"] = normal_adata.X.sum(axis=1)  # Total UMI counts
normal_adata.obs["pct_counts_mt"] = (
    normal_adata[:, normal_adata.var["mt"]].X.sum(axis=1).A1 / normal_adata.obs["total_counts"].values
) * 100
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import seaborn as sns
from umap import UMAP

# Load Latent Representations
# df_normal = pd.read_csv("normal_latent_representation.csv")
# df_tumor = pd.read_csv("tumor_latent_representation.csv")

# Combine and create labels
df_normal_latent["label"] = "Normal"
df_tumor_latent["label"] = "Tumor"
df = pd.concat([df_normal_latent, df_tumor_latent], ignore_index=True)

# Extract Features and Labels
X = df.drop(columns=["label"])  # Remove labels for clustering
y = df["label"]  # Store labels separately

# Standardize Features (important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

### K-MEANS CLUSTERING ###

# Finding optimal k using Elbow Method
inertia = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
plt.figure(figsize=(6, 4))
plt.plot(k_range, inertia, marker="o", linestyle="--")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal k")
plt.show()

# Choose optimal k based on the elbow point
optimal_k = 4  # Change based on the elbow plot

# Apply K-Means
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df["KMeans_Cluster"] = kmeans.fit_predict(X_scaled)

# Silhouette Score
silhouette_kmeans = silhouette_score(X_scaled, df["KMeans_Cluster"])
print(f"Silhouette Score for K-Means: {silhouette_kmeans:.4f}")

# Visualizing K-Means using UMAP
umap_model = UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X_scaled)

plt.figure(figsize=(7, 5))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=df["KMeans_Cluster"], palette="viridis", alpha=0.7)
plt.title("K-Means Clustering Visualization with UMAP")
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.legend(title="Cluster")
plt.show()

## DBSCAN CLUSTERING ###

# Apply DBSCAN
dbscan = DBSCAN(eps=1.0, min_samples=10)  # Adjust eps and min_samples as needed
df["DBSCAN_Cluster"] = dbscan.fit_predict(X_scaled)

# Count number of clusters (excluding noise, labeled as -1)
n_clusters_dbscan = len(set(df["DBSCAN_Cluster"])) - (1 if -1 in df["DBSCAN_Cluster"].values else 0)
print(f"DBSCAN identified {n_clusters_dbscan} clusters")

# Visualizing DBSCAN using UMAP
plt.figure(figsize=(7, 5))
sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=df["DBSCAN_Cluster"], palette="tab10", alpha=0.7)
plt.title("DBSCAN Clustering Visualization with UMAP")
plt.xlabel("UMAP1")
plt.ylabel("UMAP2")
plt.legend(title="Cluster")
plt.show()
```

```python
# Extract top 10 DEGs per cluster
top_markers_normal = degs_normal.groupby("group").head(10)
top_markers_tumor = degs_tumor.groupby("group").head(10)
```

```python
def annotate_clusters(top_markers, marker_df):
    cluster_annotation = {}
    
    for cluster, group_df in top_markers.groupby("group"):
        # Get the list of genes for this cluster
        cluster_genes = set(group_df["names"])

        # Compare with known cell markers
        best_match = None
        best_overlap = 0

        for _, row in marker_df.iterrows():
            cell_type = row["Cell Type"]
            markers = set(row["Cell Marker"])  # Known markers

            # Find overlap between cluster DEGs and known markers
            overlap = len(cluster_genes & markers)

            if overlap > best_overlap:
                best_overlap = overlap
                best_match = cell_type

        # Assign best-matching cell type
        cluster_annotation[cluster] = best_match if best_match else "Unknown"

    return cluster_annotation
```

```python
# Annotate clusters separately for normal and tumor data
normal_annotations = annotate_clusters(top_markers_normal, marker_df_clean)
tumor_annotations = annotate_clusters(top_markers_tumor, marker_df_clean)

# Map annotations back to the dataset
normal_hvg_data.obs["cell_type"] = normal_hvg_data.obs["kmeans"].astype(str).map(normal_annotations)
tumor_hvg_data.obs["cell_type"] = tumor_hvg_data.obs["kmeans"].astype(str).map(tumor_annotations)
```
